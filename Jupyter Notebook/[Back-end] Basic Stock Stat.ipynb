```python
# import libraries
import pandas as pd
import datetime
```

```python
# load data
nasdaq_df = pd.read_csv("../Data/Polished/nasdaq_2010_mid_separate_year_month_day.csv")
nasdaq_df["Date"] = pd.to_datetime(nasdaq_df["Date"])
nasdaq_df
```

```python
# mock input
'''
Input Signature:
    1. ticker symbol (string)
'''
ticker = "AAPL"
```

```python
# find the earliest year
earliest_year = nasdaq_df.loc[nasdaq_df["Ticker Symbol"] == ticker]\
    ["Year"].min()

# find the earliest month
earliest_month = nasdaq_df.loc[(nasdaq_df["Ticker Symbol"] == ticker) &\
    (nasdaq_df["Year"] == earliest_year)]\
    ["Month"].min()
```
