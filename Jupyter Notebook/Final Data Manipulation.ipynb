```python
# all necessary libraries
import pandas as pd
import random
import datetime
```

```python
# load data
jack_df = pd.read_csv("cleanedDataAllDays - Google Drive.csv")

jack_df
```

```python
# convert date column to datetime
jack_df['Date']= pd.to_datetime(jack_df['Date'], infer_datetime_format = True)
```

```python
unique_tickers = list(jack_df["Ticker Symbol"].unique())
unique_tickers.sort()

len(unique_tickers)
```

# OBJECTIVE
### A Pandas dataframe with only 1 date for each month (randomly selected) for each stock
### 1. We will first loop by individual ticker
### 2. Then loop by individual year
### 3. Then loop by individual month
### 4. Select a random date within each month

```python
# list to store all individual chosen days
all_chosen_days = []

# loop through each ticker
for ticker in unique_tickers:

    # filter the original dataframe to only the needed tickers
    lone_ticker_df = jack_df.loc[jack_df["Ticker Symbol"] == ticker]

    # get the unique years
    unique_years = list(lone_ticker_df["Year"].unique())

    # we will remove the years that are before 2011 from this date
    unique_years_from_2011 = list(lone_ticker_df["Year"].unique())
    unique_years_from_2011.sort()
    
    # remove so only 10 years are left
    for year in unique_years:
        if year < 2012:
            unique_years_from_2011.remove(year)

    # loop through each year
    for year in unique_years_from_2011:

        # filter the dataframe by the year in concern
        lone_year_df = lone_ticker_df.loc[lone_ticker_df["Year"] == year]

        # get the unique months
        unique_months = list(lone_year_df["Month"].unique())
        unique_months.sort()

        # loop through each month
        for month in unique_months:
            lone_month_df = lone_year_df.loc[lone_year_df["Month"] == month]

            # get the unique dates
            unique_days = list(lone_month_df["Day"].unique())

            # randomize a day
            chosen_day = random.choice(unique_days)

            # filter out only the chosen_day
            chosen_day_df = lone_month_df.loc[lone_month_df["Day"] == chosen_day]

            # there had been instances of duplicated rows within our dataframe, we will eliminate it by this line
            chosen_day_single_df = chosen_day_df.iloc[:1]

            # add the chosen day to the reserved list
            all_chosen_days.append(chosen_day_single_df)

# merge all chosen days into one final dataframe
final_data = pd.concat(all_chosen_days)
```

```python
# double check the final dataframe
final_data
```

```python
final_data.to_csv("randomized_day_market.csv", index = False)
```
