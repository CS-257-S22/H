```python
# import libraries
import pandas as pd
import datetime
```

```python
# load data
nasdaq_df = pd.read_csv("../Data/Polished/nasdaq_2010_mid.csv")
nasdaq_df["Date"] = pd.to_datetime(nasdaq_df["Date"])
nasdaq_df
```

```python
# extract day, month, and year value from datetime objects
day = nasdaq_df["Date"].dt.day
month = nasdaq_df["Date"].dt.month
year = nasdaq_df["Date"].dt.year
```

```python
# insert day, month, and year values back to the dataframe as separate columns
nasdaq_df.insert(1, "Year", year)
nasdaq_df.insert(1, "Month", month)
nasdaq_df.insert(1, "Day", day)

# preview
nasdaq_df
```

```python
nasdaq_df.to_csv("../Data/Polished/nasdaq_2010_mid_separate_year_month_day.csv", index = False)
```
